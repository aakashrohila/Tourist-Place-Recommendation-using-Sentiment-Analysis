```python
import pandas as pd
import numpy as np
```

```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud, STOPWORDS
```

```python
df = pd.read_csv('/kaggle/input/tourist/messages.csv')

df.head()
```

```python
df.shape
```

```python
from tqdm import tqdm, trange

corpus = ''
for i in trange(len(df)):
    corpus = corpus + df['text'][i]
    
```

```python
wordcloud = WordCloud().generate(str(corpus))

plt.imshow(wordcloud)
plt.axis("off")
plt.show()
```

```python
from matplotlib import pyplot as plt

# set width of bar
barWidth = 0.25
fig = plt.subplots(figsize =(12, 8))
 
# set height of bar
VADER = [69994 , 1333 , 4091]
ROBERTA = [69008, 574, 5839]
 
# Set position of bar on X axis
br1 = np.arange(len(VADER))
br2 = [x + barWidth for x in br1]
 
# Make the plot
plt.bar(br1, VADER, color ='r', width = barWidth,
        edgecolor ='grey', label ='VADER')
plt.bar(br2, ROBERTA, color ='g', width = barWidth,
        edgecolor ='grey', label ='ROBERTA')

# Adding Xticks
plt.xlabel('Sentiments', fontweight ='bold', fontsize = 15)
plt.ylabel('Total Count', fontweight ='bold', fontsize = 15)
plt.xticks([r + barWidth for r in range(len(VADER))],
        ['Positive' , 'Neutral' , 'Negative'])
 
plt.legend()
plt.show()
```
