```python
# Dropping unncesessary columns

df.drop(['category' , 'rating'] , axis = 1 , inplace = True)
```

```python
df

#Showing our dataframe
```

```python
df.replace({'label':{'CG':1, 'OR':0}}, inplace = True)

#Here in label, 1 represent fake data and 0 represents reviews given by a user 
```

```python
messages = df.copy()

# Here we have created copy of our dataframe
```

```python
# Data Cleaning and Preprocessing

# Importing Necessary Libraries

import re
import nltk
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('omw-1.4')

from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
```

```python
#Creating Object of WordNet Lemmatizer
wordnet = WordNetLemmatizer()
corpus = []   #Created a List named corpus.


for i in range(0,len(messages)):
    print(i)
    review = re.sub('[^a-zA-Z]' , ' ' , messages['text_'][i]) # Using Regex Function to replace digits and Special chracters with space
    review = review.lower() #converting all the reviews into lower case
    review = review.split() # splitting the reviews (by default space)
    
    # using list comprehension to compare the words which are not in stopword english library and appending them into list corpus.
    review = [wordnet.lemmatize(word) for word in review if not word in stopwords.words('english')]
    review = ' '.join(review)
    corpus.append(review)
```

```python
corpus # displaying list corpus
```

```python
df['text_'] = corpus # created a new column text_ and added the data of corpus into the column
```

```python
df.to_csv('fake_data_cleaned.csv') #converting dataframe into csv file.
```
